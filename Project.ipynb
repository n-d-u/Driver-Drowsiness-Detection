```python
import numpy as np
import cv2
from keras.models import load_model
model = load_model('yawn_detection1.h5')
from pygame import mixer
import time
mixer.init()
sound = mixer.Sound('alarm.wav')

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

# cap = cv2.VideoCapture(0) #for webcam input

cap = cv2.VideoCapture('awake.mp4')

score = 0

while 1:
    ret, img = cap.read()
    height,width = img.shape[:2]
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    # img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        
        eyes = eye_cascade.detectMultiScale(roi_gray)
        #for (ex,ey,ew,eh) in eyes:
        #    cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        #    crop = img[ex:ex+ew, ey:ey+eh]
        #    cv2.imshow('img',crop)   

        i = 0
        per = []
        max = 0
        maxper = 0
        for (ex,ey,ew,eh) in eyes:
            per = ex+ey+ew+eh
            if per>maxper:
                max = i
                maxper = per
            i += 1
        x,y,w,h = eyes[max]
        crop = img[x:x+w, y:y+h]
        eyes_roi = roi_gray[y: y+h, x:x + w]
        eyes_roi = cv2.resize(eyes_roi, (256,256))
        img_batch = np.expand_dims(eyes_roi, 0) 
        # img_batch = cv2.cvtColor(img_batch, cv2.COLOR_RGB2GRAY)
        # print(model.predict(img_batch))
        y_prob = model.predict(img_batch) 
        y_classes = y_prob.argmax(axis=-1)
        # print(y_classes)
        
        if (y_classes[0] == 0):
            score=score+1
            cv2.putText(img,"Closed",(10,height-20), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1,(255,255,255),1,cv2.LINE_AA)
            cv2.putText(img,'Score:'+str(score),(100,height-20), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1,(255,255,255),1,cv2.LINE_AA)
        else:
            score = score - 5
            if (score<0):
                score = 0
            cv2.putText(img,"Open",(10,height-20), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1,(255,255,255),1,cv2.LINE_AA)
            cv2.putText(img,'Score:'+str(score),(100,height-20), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1,(255,255,255),1,cv2.LINE_AA)

        if(score<0):
            score=0
            cv2.putText(img,'Score:'+str(score),(100,height-20), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1,(255,255,255),1,cv2.LINE_AA)
        
        if(score>15):
            try:
                sound.play()
            except: # isplaying = False
                pass
            
        
    
    cv2.imshow('img',img)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
